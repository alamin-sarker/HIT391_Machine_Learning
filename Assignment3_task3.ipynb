##Importing Libraries

```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from sklearn.metrics import pairwise_distances

nltk.download("punkt")
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')

from nltk import word_tokenize
```

## 1. Data Exploration

```python
# Import the question file" questions"
df= pd.read_csv("/content/drive/MyDrive/ML_Assignments/questions.csv")
# The first 5 rows of our DataFrame
df.head()
```

## 2. Data Preprocessing
Applying normalizaiton to the data

```python
# Function that performs text normalization steps

def text_normalization(text):
    text=str(text).lower()
    spl_rev_text=re.sub(r'[^ a-z]','',text)
    tokens=nltk.word_tokenize(spl_rev_text)
    lemma=WordNetLemmatizer()
    tags_list=pos_tag(tokens,tagset=None)
    lemma_words=[]
    for token,pos_token in tags_list:
        if pos_token.startswith('V'):
            pos_val='v'
        elif pos_token.startswith('J'):
            pos_val='a'
        elif pos_token.startswith('R'):
            pos_val='r'
        else:
            pos_val='n'
        lemma_token=lemma.lemmatize(token,pos_val)
        lemma_words.append(lemma_token)

    return " ".join(lemma_words)
```

```python
# Applying the function to the dataset to get clean text
df['lemmatized_text']=df['Questions'].apply(text_normalization)
df
```

```python
#Examine all the stop words of the data

stop = stopwords.words('english')
print(stop)
```

##3. Testing Bag of Words (BOW) approach

*   BOW
*   Cosine Similarity



```python
# Intializing the count vectorizer
cv = CountVectorizer()
X = cv.fit_transform(df['lemmatized_text']).toarray()
```

```python
# Returns all the unique words from data
features = cv.get_feature_names_out()
df_bow = pd.DataFrame(X, columns = features)
df_bow.head()
```

```python
# Considering an example query
Question1 ='What is the dataset in task 1 about?'
```

```python
# Checking for stop words
Q=[]
a=Question1.split()
for i in a:
    if i in stop:
        continue
    else:
        Q.append(i)
    b=" ".join(Q)
```

```python
# Applying the text normalization function and BOW
Question1_lemma = text_normalization(b)
Question1_bow = cv.transform([Question1_lemma]).toarray()
```

```python
text_normalization
```

```python
Question1_bow
```

```python
# Cosine similarity for the above question we considered.

cosine_value = 1- pairwise_distances(df_bow, Question1_bow, metric = 'cosine' )
(cosine_value)
```

```python
# Creating a new column
df['similarity_bow']=cosine_value
```

```python
# Taking the similarity value of answers for the question
df_simil = pd.DataFrame(df, columns=['Answers','similarity_bow'])
df_simil
```

```python
# Sorting the similarity values
df_simil_sort = df_simil.sort_values(by='similarity_bow', ascending=False)
df_simil_sort.head()
```

```python
# Considering the value of similiarity to be greater than 0.2
threshold = 0.2
df_threshold = df_simil_sort[df_simil_sort['similarity_bow'] > threshold]
df_threshold
```

Finally using bow for the question 'What is the dataset in task 1 about?' , we consider the response with the highest similarity.

```python
# Returns the index number of highest value
index_value = cosine_value.argmax()
index_value
```

```python
(Question1)
```

```python
# The text at the above index becomes the answer for the question
df['Answers'].loc[index_value]
```

##5. Chatbot with model of Bag of Words

```python
# Function that removes stop words and processes the text
def stopword_(text):
    tag_list=pos_tag(nltk.word_tokenize(text),tagset=None)
    stop=stopwords.words('english')
    lemma=WordNetLemmatizer()
    lemma_word=[]
    for token,pos_token in tag_list:
        if token in stop:
            continue
        if pos_token.startswith('V'):
            pos_val='v'
        elif pos_token.startswith('J'):
            pos_val='a'
        elif pos_token.startswith('R'):
            pos_val='r'
        else:
            pos_val='n'
        lemma_token=lemma.lemmatize(token,pos_val)
        lemma_word.append(lemma_token)
    return " ".join(lemma_word)
```

```python
# Defining a function that returns the answer to query using bow
def chat_bot1(text):
    s=stopword_(text)
    lemma=text_normalization(s)
    bow=cv.transform([lemma]).toarray()
    cosine_value = 1- pairwise_distances(df_bow,bow, metric = 'cosine' )
    index_value=cosine_value.argmax()
    return df['Answers'].loc[index_value]
```
